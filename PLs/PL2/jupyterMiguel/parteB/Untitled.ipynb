```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

```python
def linear(x,m,b):
    return m * x + b
```

```python
df = pd.read_csv("parteB.csv", delimiter = "\t")

lambdas= np.array(df["lambda"])
Vcs= np.array(df["V"])

ls = np.zeros(9)
ls[:5] = lambdas[:5]
ls[5:] = lambdas[6:]
Vs = np.zeros(9)
Vs[:5] = Vcs[:5]
Vs[5:] = Vcs[6:]

for i in range(9):
    print(ls[i],Vs[i])
    
mus = 299792458/ (ls*(10**-9))
```

```python
params, mc= curve_fit(linear, mus, Vs)

plt.title('Gráfico 12: Medidas efetuadas com condensador para\nmedição de potenciais de paragem')
plt.scatter(mus,Vs, label= 'Dados obtidos experimentalmente')
plt.plot(mus, mus*params[0]+params[1], color = "r", label = 'Ajuste linear')
plt.grid()
plt.ylabel("$V_c$ / V")
plt.xlabel(r"$\nu$ / Hz")

plt.legend()
plt.savefig('graph12')
plt.show()
```

```python
print( params[0], np.sqrt(mc[0,0]) *3, '\n' )
print( params[1], np.sqrt(mc[1,1]) *3 )
```
