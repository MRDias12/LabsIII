```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp
from sklearn.metrics import r2_score
from scipy.signal import find_peaks
```

```python
def top_peaks(df, intensity_offset=0, intensity_col="Intensity", wavelength_col="Wavelength", 
              top_n=20, prominence=50):
    
    y = (df[intensity_col] - intensity_offset).values
    x = df[wavelength_col].values

    peaks, props = find_peaks(y, prominence=prominence)
    
    peaks_df = pd.DataFrame({
        "Wavelength (nm)": x[peaks],
        "Contagens": y[peaks]
    })
    
    #peaks_df = peaks_df.sort_values("Intensity", ascending=False).head(top_n).reset_index(drop=True)
    return peaks_df
```

```python
def exponencial(x, a, b, c):
    return a * np.exp(- x / b) + c
```

```python
taus = np.zeros(13)
for i in range(2, 15):
    data = pd.read_csv(f"Flu{i + 1}.csv")
    max_idx = np.argmax(data["Y"]) + 1
    t_max = data["time"][max_idx]
    final_idx = np.argmin(np.abs(data["time"] - 0.25))
    popt, pcov = sp.optimize.curve_fit(exponencial, 
                                       data["time"][max_idx:final_idx] - t_max, data["Y"][max_idx:final_idx],
                                      [1, 0.01, 0.5])
    print(popt[1])
    print(np.sqrt(np.diag(pcov)[1]))

    taus[i - 2] = popt[1]

    plt.plot(data["time"], data["Y"], label="Dados experimentais")
    plt.plot(data["time"][max_idx:final_idx], 
             exponencial(data["time"][max_idx:final_idx] - t_max, popt[0], popt[1], popt[2]),
            label="Ajuste exponencial")
    plt.xlim(0.225, 0.25)
    plt.grid()
    plt.xlabel("t/s")
    plt.ylabel("I")
    plt.title(f"Medio {i + 1} da intensidade da radiao fluorescente")
    plt.legend()
    plt.show()
    print("\n")
```

```python
print(taus, "\n")
print(np.mean(taus))
print(np.std(taus, ddof=1) / np.sqrt(len(taus)))
```
